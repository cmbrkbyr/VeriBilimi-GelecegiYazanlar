# Basit Dogrusal Regresyon

### model

```python
import pandas as pd
```

```python
df =pd.read_csv("Advertising.csv")
df =df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.jointplot(x="TV", y="sales",data=df,kind="reg");
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["TV"]]
X.head()
```

```python
y = df[["sales"]]
```

```python
reg = LinearRegression()
```

```python
model=reg.fit(X,y)
```

```python
model.intercept_ # B0 degeri
```

```python
model.coef_ # b1 degeri
```

```python
#rkare
model.score(X,y)
```

### tahmin

```python
import matplotlib.pyplot as plt
```

```python
g = sns.regplot(df["TV"],df["sales"],ci=None,scatter_kws={'color':'r','s':9})
g.set_title("Model Denklemi: Sales = 7.03 + TV*0.05")
g.set_ylabel("Satis Sayisi")
g.set_xlabel("TV Harcamalari")
plt.xlim(-10,310)
plt.ylim(bottom=0);
```

## Artiklar ve Makine grenmesindeki nemi

### MSE : Hata Karaler Ortalamasi
### RMSE : Hata Kareler Ortalamasinin Karekk

```python
gercek_y= y[0:10]
```

```python
tahmin_edilen_y=pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar = pd.concat([gercek_y,tahmin_edilen_y],axis=1)
hatalar.columns=["gercek_y","tahmin_edilen_y"]
hatalar
```

```python
hatalar["hata"]= hatalar["gercek_y"]-hatalar["tahmin_edilen_y"]
```

```python
hatalar["hata_kareler"] = hatalar["hata"]**2
```

```python
hatalar
```

```python
import numpy as np
np.mean(hatalar["hata_kareler"])
```
